```python
def f(x):
    return (x+3)**2
def df(x):
    return 2*x+6
def gradientdesc(initial_x, learning_rate, num_iterations):
    x=initial_x
    x_history=[x]
    for i in range(num_iterations):
        gradient =df(x)
        x=x-learning_rate*gradient
        x_history.append(x)
    return x, x_history

initial_x=2
learning_rate=.01
num_iterations=90

x, x_history=gradientdesc(initial_x,learning_rate, num_iterations)

print("Local minima{}".format(x))
for i in range(len(x_history)):
    print(f"Iterator number{i}:{x_history[i]}")
    
    
```

```python
x_vals=np.linspace(-11,5,50)
plt.plot(x_vals, f(x_vals))
plt.plot(x_history, f(np.array(x_history)),'rx')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
data=pd.read_csv('sales_data_sample.csv',encoding='ISO-8859-1')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
x=data[['QUANTITYORDERED','PRICEEACH']].values
```

```python
x
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cluster import KMeans
inertia=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), inertia, marker='o', linestyle='-',color='r')
plt.xlabel('No.of clusters')
plt.ylabel('Inertia')
```

```python
opti=3
kmeans=KMeans(n_clusters=opti, random_state=0)
data['CLUSTER']=kmeans.fit_predict(x)

print(data.head)
```
