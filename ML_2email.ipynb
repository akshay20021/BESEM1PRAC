```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("emails.csv")
```

```python
df=df.dropna()
```

```python
df.columns
```

```python
x=df.iloc[: ,1:3001]
x
```

```python
y=df.iloc[:,3001]
y
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
```

```python
kn=KNeighborsClassifier()
kn.fit(x_train,y_train)
```

```python
sc=SVC()
sc.fit(x_train,y_train)
```

```python
ky_pred=kn.predict(x_test)
sy_pred=sc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
acc=accuracy_score(y_test,ky_pred)
pscore=precision_score(y_test,ky_pred)
re_score=recall_score(y_test,ky_pred)
f1=f1_score(y_test,ky_pred)
```

```python
sacc=accuracy_score(y_test,sy_pred)
spscore=precision_score(y_test,sy_pred)
sre_score=recall_score(y_test,sy_pred)
sf1=f1_score(y_test,sy_pred)
```

```python
print("Accuracy of KNN:{}".format(acc))
```

```python
print("Accuracy of SVM:{}".format(sacc))
```

```python
mat=confusion_matrix(y_test,ky_pred)
mat
```

```python
sns.heatmap(mat, annot=True, fmt='d', cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
